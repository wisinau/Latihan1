## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Pola Pembelian Pelanggan?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Pastikan order_purchase_timestamp dalam format datetime
df_orders_clean["order_purchase_timestamp"] = pd.to_datetime(df_orders_clean["order_purchase_timestamp"])

# Menambahkan kolom bulan dari timestamp pembelian
df_orders_clean["order_month"] = df_orders_clean["order_purchase_timestamp"].dt.to_period("M")

# Mengelompokkan data berdasarkan bulan dan menghitung jumlah pesanan
monthly_sales = df_orders_clean.groupby("order_month").size()

# Visualisasi tren penjualan di seluruh Brasil
plt.figure(figsize=(10, 6))
monthly_sales.plot()
plt.title("Overall Sales Trends in Brazil Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

**Insight:**
- Jumlah pesanan meningkat signifikan dari akhir 2016 hingga awal 2018, menunjukkan pertumbuhan bisnis yang pesat.  
- Terjadi lonjakan besar sekitar Oktober 2017, kemungkinan akibat promosi, musim belanja, atau faktor lain.  
- Setelah mencapai puncak pada akhir 2017, jumlah pesanan mengalami fluktuasi tetapi tetap tinggi dibandingkan awal periode.  
- Setelah pertengahan 2018, tren mulai menunjukkan pola yang lebih stabil.

### Pertanyaan 5: Bagaimana Pola Loyalitas Pelanggan Berdasarkan RFM Analysis?

```python
from datetime import datetime

# Konversi tanggal ke datetime
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])

# Tentukan tanggal referensi (misalnya, hari terakhir transaksi)
reference_date = df_orders['order_purchase_timestamp'].max()

# Hitung Recency (hari sejak transaksi terakhir)
rfm = df_orders.groupby('customer_id').agg(
    recency=('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    frequency=('order_id', 'count')
).reset_index()

# Gabungkan dengan order_payments untuk menghitung Monetary
df_pay_agg = df_pay.groupby('order_id')['payment_value'].sum().reset_index()
df_orders_pay = df_orders.merge(df_pay_agg, on='order_id', how='left')

rfm = rfm.merge(df_orders_pay.groupby('customer_id')['payment_value'].sum().reset_index(), on='customer_id', how='left')
rfm.rename(columns={'payment_value': 'monetary'}, inplace=True)

# Tampilkan hasil
rfm.head()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(rfm['recency'], bins=30, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Distribusi Recency')
axes[0].set_xlabel('Hari sejak transaksi terakhir')

sns.histplot(rfm['frequency'], bins=30, kde=True, ax=axes[1], color='green')
axes[1].set_title('Distribusi Frequency')
axes[1].set_xlabel('Jumlah Transaksi')

sns.histplot(rfm['monetary'], bins=30, kde=True, ax=axes[2], color='red')
axes[2].set_title('Distribusi Monetary')
axes[2].set_xlabel('Total Pengeluaran')

plt.show()
```

**Insight:**  

**Distribusi Recency (Hari sejak transaksi terakhir):**  
   - Mayoritas pelanggan melakukan transaksi terakhir dalam 0-400 hari terakhir.  
   - Ada pola distribusi yang cukup merata, namun jumlah pelanggan mulai menurun setelah 400 hari.  
   - Ini menunjukkan ada pelanggan yang masih cukup aktif, tetapi ada juga yang sudah lama tidak bertransaksi.  

**Distribusi Frequency (Jumlah transaksi):**  
   - Hampir semua pelanggan hanya memiliki **1 transaksi**, yang menunjukkan pola pembelian one-time purchase.  
   - Ini bisa menjadi indikasi bahwa banyak pelanggan tidak kembali setelah pembelian pertama.  
   - Program loyalitas atau retargeting bisa diperlukan untuk meningkatkan repeat purchase.  

**Distribusi Monetary (Total Pengeluaran):**  
   - Sebagian besar pelanggan memiliki total pengeluaran yang rendah, sementara hanya sedikit pelanggan yang memiliki pengeluaran sangat tinggi.  
   - Ini menunjukkan adanya pelanggan high-value yang mungkin perlu mendapatkan perlakuan khusus, seperti promosi eksklusif atau program VIP.
